```python
# y = sin(x) is a beautiful curve. 
# It's a periodic function with period 2Ï€. 
```

```python
import numpy as np
import matplotlib.pyplot as plot
from matplotlib.pyplot import figure

time = np.arange(-1, 10, 0.1)
amplitude = np.sin(time)
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot.plot(time, amplitude)

plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```python
# Say I want to generate sine wave with a higher rate or a slower rate
# I'll just multiply a constant (say 'r') to the sin(x)

# Making my formula look like :    y = r * (sin(x))

# if r == 1, we get the above plot. But what will happen if we increase or decrease the rate?
```

```python
# let's increase it for sometime
r = 2
time = np.arange(-1, 10, 0.1)
amplitude = r * np.sin(time)
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot.plot(time, amplitude)

plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```python
# Okay. Interesting! The curve maintained it's shape. It's just the amplitude that got increased. 
# Overlapping both the curve will make it noticable.
```

```python
r_high = 2
r_low = 0.5

time = np.arange(-1, 10, 0.1)
amplitude = np.sin(time)
amplitude_high = r_high * np.sin(time)
amplitude_low = r_low * np.sin(time)

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot.plot(time, amplitude, color='red')
plot.plot(time, amplitude_high, color='orange')
plot.plot(time, amplitude_low, color='blue')

plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```python
# It has already started getting beautiful. Looks like a smiling face. But where's the chaos? 
# Isin't smiling face a chaos for many? Kidding. 
# The red curve denotes r = 1, for blue r = 0.5, for orange r = 2

#It very visible that with increasing the constant r, curve only changes the amplitude

# the equation sin(t), with varying t can be written as sin(x(n)) where x(n) at a given n will give the 
# time at that instance.

# Hence modifying out equation as :  y = r*sin(x(n))

# Just to let you visualise what x(n) will look like
```

```python
time = np.arange(-1, 10, 0.1)

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot.plot(time, time, color='orange')

plot.xlabel('x')
plot.ylabel('Time')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```python
r_high = 2
r_low = 0.5

time = np.arange(-1, 10, 0.1)
amplitude = np.sin(time)
amplitude_high = r_high * np.sin(time)
amplitude_low = r_low * np.sin(time)

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot.plot(amplitude, time, color='red')
plot.plot(amplitude_high, time, color='orange')
plot.plot(amplitude_low, time, color='blue')

plot.xlabel('Amplitude')
plot.ylabel('Time')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```python
# There you go, you get a beautiful river flowing. The blue and the red curve contains the river water. 
# The orange curve will tell a rough estimate of what the river would look like after few years. The corrosion of
# land starts taking place along the direction of orange curve

# Too many things to be derived from this graph but we are searching for a chaos. This isin't chaos.

# What would happen if we try to equate this expression against x(n+1)
# Then our equation becomes x(n+1) = r * sin(x(n))
```

```python
r = 1

x_n = np.arange(-1, 10, 0.1)
x_n_1 = x_n + 1

amplitude = r * np.sin(x_n)

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot.plot(x_n_1, amplitude)

plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```python
# Just a graph that's rotated by some angle

# Let's try to vary the rate or our cute small constant 'r' between [0, 1] and plot againt the varying constant
```

```python
r = np.arange(0, 10, 0.1)

x_n = np.arange(0, 10, 0.1)

amplitude = (r * np.sin(x_n))
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot.plot(r, amplitude, color='orange')

plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
```

```python
# We get an unbounded wave, it's amplitude will go on increasing and one day, Boom. Will cover the whole universe :P 
```

```python
# What will happen if we try to equate the curve with it's future input value i.e. x(n+1)

# the equation becomes          x(n+1) = r * sin(x(n))

# this results in a logistic equation. For a given x(n) we want to compute it's value at x(n+1)
#Lets store the values for different r

# To visualise this, let's use a logistic map
```

```python
import math

def logistic_iteration(x, r):
    return r * math.sin(x)
```

```python
r = np.linspace(0, 2, num=400)
x = np.linspace(0, 2, num=400)
array = np.empty((r.size, x.size))
for i in range(array.shape[0]):
    for j in range(array.shape[1]):
        def value(x, r, iters=10):
            for _ in range(iters):
                x = logistic_iteration(x, r)
            return x
        array[i, j] = value(x[j], r[i])
```

```python
figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
plot.plot([rr * np.ones(array.shape[0]) for rr in r], [row for row in array], 'x', ms=0.9)
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.xlabel('rate (r)')
plot.ylabel('Equilibrium ')
plot.show()
```

```python
# Whoa. That's something interesting. But we do not notice anything unusual. The curve is rising to a value, 
# it's falling for a small space, then again it starts rising.

# no chaos. 

# Let's increase out value of r to 3 and see
```

```python
r = np.linspace(0, 3, num=400)
x = np.linspace(0, 2, num=400)
array = np.empty((r.size, x.size))
for i in range(array.shape[0]):
    for j in range(array.shape[1]):
        def value(x, r, iters=10):
            for _ in range(iters):
                x = logistic_iteration(x, r)
            return x
        array[i, j] = value(x[j], r[i])
```

```python
figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
plot.plot([rr * np.ones(array.shape[0]) for rr in r], [row for row in array], 'x', ms=0.9)
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.xlabel('rate (r)')
plot.ylabel('Equilibrium ')
plot.show()
```

```python
# Beautiful 

# Somewhere around 2.1, we see the graph starts to splits into two parts. The value at that point could be anything
# between the two. 
# It doesn't stops there. Somewhere around 2.5 each split end splits again. Now at any given time, the graph has
# 4 values to offer. A complete mess.

# Let's increase the value of r further to let's say 5
```

```python
r = np.linspace(0, 5, num=400)
x = np.linspace(0, 2, num=400)
array = np.empty((r.size, x.size))
for i in range(array.shape[0]):
    for j in range(array.shape[1]):
        def value(x, r, iters=10):
            for _ in range(iters):
                x = logistic_iteration(x, r)
            return x
        array[i, j] = value(x[j], r[i])
```

```python
figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
plot.plot([rr * np.ones(array.shape[0]) for rr in r], [row for row in array], 'x', ms=0.9)
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.xlabel('rate (r)')
plot.ylabel('Equilibrium ')
plot.show()
```

```python
# The Graph literally blasted. It's just a CHAOS after rate goes above 3.
# There's a quote which i'd like to mention here 

# "Chaos theory simply suggests that what appears to most people as chaos is not really chaotic, 
#  but a series of different types of orders with which the human mind has not yet become familiar." 
#  ~ Frederick Lenz

# You can trace the whole curve if the initial conditions are same. You can exactly trace it. But a tiniest deviation
# will not be noticeable at start but will lead to a totally different world later on.

# This is the science behind the butterfly effect which goes like "A butterfly flaps its wings in the Amazonian 
# jungle, and subsequently a storm ravages half of Europe."
```
